## Generalist vs. Specialist Models

```python
# Generalist models are trained on broad, diverse datasets
# Examples: GPT-4, Claude, LLaMA base models

# Specialist models are fine-tuned for specific tasks/domains
# Examples: CodeLlama, Med-PaLM, domain-specific fine-tunes

print("Generalist models: Broad capabilities, less specialized")
print("Specialist models: Focused capabilities, domain expertise")
```

## When to Fine-Tune

```python
# Decision matrix for fine-tuning
scenarios = [
    {"scenario": "Domain-specific knowledge", "fine_tune": True, "reason": "Requires specialized terminology"},
    {"scenario": "Simple tasks with clear instructions", "fine_tune": False, "reason": "Prompt engineering sufficient"},
    {"scenario": "Specific output format needed", "fine_tune": True, "reason": "Requires consistent formatting"},
    {"scenario": "Knowledge-intensive tasks", "fine_tune": False, "reason": "RAG may be better"},
]

for scenario in scenarios:
    print(f"\nScenario: {scenario['scenario']}")
    print(f"Fine-tune: {scenario['fine_tune']}")
    print(f"Reason: {scenario['reason']}")
```

## Hardware Requirements Estimation

```python
import torch

# Memory requirements for different model sizes
def calculate_memory(model_size_billions, method="full"):
    """Calculate GPU memory requirements"""
    if method == "full":
        return model_size_billions * 2  # FP16
    elif method == "lora":
        return model_size_billions * 2  # Same as full for base model
    elif method == "qlora":
        return model_size_billions * 0.8  # 4-bit quantization

model_sizes = [7, 13, 34, 70]
methods = ["full", "lora", "qlora"]

print("Memory Requirements (GB):")
print("\nModel Size | Full | LoRA | QLoRA")
print("-" * 35)
for size in model_sizes:
    memories = [calculate_memory(size, method) for method in methods]
    print(f"{size:9}B | {memories[0]:4.0f} | {memories[1]:4.0f} | {memories[2]:4.0f}")
```

## Cost Estimation

```python
# Cloud training cost calculator
def estimate_training_cost(model_size, training_hours, gpu_hourly_rate):
    """Estimate cloud training cost"""
    return training_hours * gpu_hourly_rate

# Example: 7B model with LoRA
model_size = 7
training_hours = 4
gpu_rate = 1.00  # $1.00/hour

cost = estimate_training_cost(model_size, training_hours, gpu_rate)
print(f"Estimated training cost: ${cost:.2f}")
```
