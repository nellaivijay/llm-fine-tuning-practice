# Chapter 4: Data Preparation

This notebook covers data collection, cleaning, and formatting for fine-tuning.

## Data Loading

```python
import pandas as pd
from datasets import Dataset

# Create sample data
sample_data = [
    {"input": "What is machine learning?", "output": "Machine learning is a subset of AI..."},
    {"input": "Explain neural networks", "output": "Neural networks are computing systems..."},
    {"input": "What is deep learning?", "output": "Deep learning uses neural networks..."},
]

df = pd.DataFrame(sample_data)
dataset = Dataset.from_pandas(df)
print(f"Dataset size: {len(dataset)}")
```

## Data Cleaning

```python
# Remove duplicates
df = df.drop_duplicates()

# Remove empty entries
df = df[df['input'].str.strip() != '']
df = df[df['output'].str.strip() != '']

# Filter by length
df['input_length'] = df['input'].str.len()
df = df[(df['input_length'] > 10) & (df['input_length'] < 1000)]

print(f"Cleaned dataset size: {len(df)}")
```

## Synthetic Data Generation

```python
templates = [
    "What is {topic}?",
    "Explain {topic} in simple terms.",
    "How does {topic} work?",
]

topics = ["machine learning", "neural networks", "deep learning", "AI"]

synthetic_data = []
for template in templates:
    for topic in topics:
        synthetic_data.append({"input": template.format(topic=topic), "output": ""})

print(f"Generated {len(synthetic_data)} synthetic examples")
```

## Tokenization

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token

# Tokenize sample
text = "What is machine learning?"
tokens = tokenizer(text, return_tensors="pt")

print(f"Tokenized: {tokens}")
print(f"Token count: {len(tokens['input_ids'][0])}")
```

## Key Takeaways

1. Data quality is crucial for fine-tuning
2. Clean and validate data before training
3. Synthetic data can augment limited datasets
4. Proper formatting is essential
5. Tokenization converts text to model input
