## CLIP Example (Image-Text Retrieval)

```python
# CLIP-based retrieval (conceptual)
import torch
import torch.nn as nn

# Simulate CLIP embeddings
image_features = torch.randn(1, 512)  # Image embedding
text_features = torch.randn(3, 512)   # Text embeddings for 3 labels

# Calculate similarity
similarity = (image_features @ text_features.T).softmax(dim=-1)

labels = ["dog", "cat", "bird"]
predicted_label = labels[similarity.argmax()]

print(f"Similarities: {similarity[0].tolist()}")
print(f"Predicted label: {predicted_label}")
```

## VLM Evaluation Metrics

```python
# VLM-specific metrics
def recall_at_k(retrieved_items, relevant_items, k):
    """Calculate Recall@K"""
    retrieved_k = retrieved_items[:k]
    relevant_retrieved = len(set(retrieved_k) & set(relevant_items))
    return relevant_retrieved / len(relevant_items)

def mean_reciprocal_rank(retrieved_items, relevant_items):
    """Calculate MRR"""
    for i, item in enumerate(retrieved_items):
        if item in relevant_items:
            return 1 / (i + 1)
    return 0

# Example
retrieved = ["dog", "cat", "bird", "car"]
relevant = ["cat", "bird"]

recall_1 = recall_at_k(retrieved, relevant, 1)
recall_2 = recall_at_k(retrieved, relevant, 2)
mrr = mean_reciprocal_rank(retrieved, relevant)

print(f"Recall@1: {recall_1:.2f}")
print(f"Recall@2: {recall_2:.2f}")
print(f"MRR: {mrr:.2f}")
```
